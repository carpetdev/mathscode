```python
n = 3
X = list(range(n))
counts = {"magma": n^n^2, "semigroup": 0}

def table_generator():
    return (tuple(l[i: i+n] for i in range(0, n^2, n)) for l in itertools.product(*[X]*n^2))

tables = table_generator()
# percents = [0]
for i, table in enumerate(tables):
    is_associative = True
    for a,b,c in itertools.product(*[X]*3):
        if table[a][table[b][c]] != table[table[a][b]][c]:
            is_associative = False
            break
    counts["semigroup"] += is_associative
    # percent = i / counts["magma"]
    # if percent - percents[-1] >= 0.1:
    #     print(round(percent*100), "% done")
    #     percents.append(percent)
```
